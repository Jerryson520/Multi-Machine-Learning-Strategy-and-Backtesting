```python
from alpaca.trading.client import TradingClient
trading_client = TradingClient(api_key, api_secret, paper=True)
account = trading_client.get_account()
print(account.account_number)
```

## Collect all assets (Crypto)

```python
from alpaca.trading.requests import GetAssetsRequest
from alpaca.trading.enums import AssetClass
trading_client = TradingClient(api_key, api_secret, paper=True) # search for crypto assets
search_params = GetAssetsRequest(asset_class=AssetClass.CRYPTO)
assets = trading_client.get_all_assets(search_params)
for asset in assets:
    print(asset)
```

## Submit Order

```python
from alpaca.trading.requests import MarketOrderRequest
from alpaca.trading.enums import OrderSide, TimeInForce
trading_client = TradingClient(api_key, api_secret, paper=True)# preparing orders
market_order_data = MarketOrderRequest(
                   symbol="AAPL",
                   qty=0.1,
                   side=OrderSide.BUY,
                   time_in_force=TimeInForce.DAY
                   ) # Market order
market_order = trading_client.submit_order(
               order_data=market_order_data
              )
print(market_order)
```

```python
from alpaca.trading.requests import GetOrdersRequest
from alpaca.trading.enums import OrderSide, OrderStatus
trading_client = TradingClient(api_key, api_secret, paper=True) # params to filter orders by
request_params = GetOrdersRequest(
                    status='all',
                    side=OrderSide.SELL
                 ) # orders that satisfy params
orders = trading_client.get_orders(filter=request_params)
for order in orders:
    print(order)
```

## Cancel order

```python
trading_client = TradingClient(api_key, api_secret, paper=True) # Attempt to cancel order
cancel_status = trading_client.cancel_order_by_id(orders[0].model_dump()['id'].hex)
print(cancel_status)
```
